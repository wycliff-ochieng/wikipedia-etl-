```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://en.wikipedia.org/wiki/List_of_largest_universities")
```

```python
#response.content
```

```python
soup = BeautifulSoup(response.content)
```

```python
soup.findAll('table')
```

```python
table = soup.find('table', attrs={'class':'sortable'})
```

```python
trs = table.findAll('tr')
```

```python
trs
```

```python
for item in trs[0].findAll('th'):
    print(item.text.strip())
```

```python
#OR
columns = []
for item in trs[0].findAll('th'):
    print(item.text.strip())
    columns.append(item.text.strip())
```

```python
columns[-1] = "Link"
```

```python
columns
```

```python
list(map(lambda x: x.text.strip(),trs[0].findAll('th')))
```

```python
rows = trs[1:]
rows[0]
```

```python
#extract link for each row
rows[0].findAll('td')[1].a.attrs['title']
```

```python
#extraction from a singe row
def extract_row(tr):
    row_soup_list = tr.findAll('td')
    row = list(map(lambda x:x.text.strip(),row_soup_list))
    if len(row)<6:
        row.append('')
        print("=========================")
        print(row)
    link = row_soup_list[1].a.attrs['href']
    #print('Link:', link)
    row[-1] = f'https://en.wikipedia.org{link}' if link else ''
    return row
```

```python
extract_row(rows[1])
```

```python
data = list(map(extract_row,rows))
```

```python
data
```

```python
import csv
import json
import pandas as pd
```

```python
#save to csv file
with open('Universities.csv','w', newline='') as f:
    csvwriter = csv.writer(f)
    csvwriter.writerow(columns)
    csvwriter.writerows(data)
```

```python
#save to json file
with open('Universities.json', 'w') as f:
    json.dump(data, f, indent=4)
```

```python
#put data to a dataframe
df = pd.DataFrame(data=data,columns=columns )
```
